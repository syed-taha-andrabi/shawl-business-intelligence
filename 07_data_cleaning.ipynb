```python
import pandas as pd
import os

print(os.getcwd())
print(os.listdir('data/'))
```

```python
df = pd.read_csv('data/amazon_data.csv')
print(df.shape)
print(df.columns.tolist())
print(df.head())
```

```python
df_clean = df[['data', 'data2', 'data5', 'price', 'image']].copy()

df_clean.columns = ['brand', 'title', 'reviews', 'price', 'image']

print(df_clean.head())
print(df_clean.shape)
```

```python
# drop rows where price is missing
df_clean = df_clean.dropna(subset=['price'])

# clean reviews column - remove brackets and convert
df_clean['reviews'] = df_clean['reviews'].str.replace('(', '').str.replace(')', '').str.replace('K', '000').str.strip()
df_clean['reviews'] = pd.to_numeric(df_clean['reviews'], errors='coerce').fillna(0).astype(int)

# reset index
df_clean = df_clean.reset_index(drop=True)

print(df_clean.head(10))
print(f"\nTotal clean products: {len(df_clean)}")
print(f"\nPrice range: AED {df_clean['price'].min()} — AED {df_clean['price'].max()}")
print(f"Average price: AED {df_clean['price'].mean():.2f}")
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# price distribution
axes[0].hist(df_clean['price'], bins=10, color='steelblue', edgecolor='black')
axes[0].set_title('Competitor Price Distribution (Amazon.ae)')
axes[0].set_xlabel('Price (AED)')
axes[0].set_ylabel('Number of Products')

# top brands by average price
brand_avg = df_clean.groupby('brand')['price'].mean().sort_values(ascending=False).head(8)
axes[1].barh(brand_avg.index, brand_avg.values, color='coral')
axes[1].set_title('Average Price by Brand')
axes[1].set_xlabel('Average Price (AED)')

plt.tight_layout()
plt.show()
```

```python
df_clean.to_csv('data/amazon_clean.csv', index=False)
print("Saved!")
```
