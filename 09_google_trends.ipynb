```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt
import time

# connect to Google Trends
pytrends = TrendReq(hl='en-US', tz=360)
print("Connected!")
```

```python
# search terms related to your father's business

keywords = ['Pashmina shawl', 'Kashmiri shawl', 'Kani shawl', 'luxury shawl']

pytrends.build_payload(keywords, timeframe='today 5-y', geo='')

df_trends = pytrends.interest_over_time()
df_trends = df_trends.drop(columns=['isPartial'])

print(df_trends.head())
print(f"\nShape: {df_trends.shape}")
```

```python
time.sleep(15)
plt.figure(figsize=(14, 6))

for col in df_trends.columns:
    plt.plot(df_trends.index, df_trends[col], label=col, marker='')

plt.title('Search Interest Over 5 Years — Kashmiri Shawl Keywords')
plt.xlabel('Date')
plt.ylabel('Search Interest (0-100)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
