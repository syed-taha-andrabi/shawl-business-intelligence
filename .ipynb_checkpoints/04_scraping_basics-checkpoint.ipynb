```python
import requests
from bs4 import BeautifulSoup

url = "http://books.toscrape.com"
response = requests.get(url)

print(response.status_code)
print(type(response.text))
print(response.text[:500])
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.title)
```

```python
books = soup.find_all('article', class_='product_pod')
print(f"Found {len(books)} books on this page")

for book in books[:20]:  # first 5 only
    title = book.find('h3').find('a')['title']
    price = book.find('p', class_='price_color').text
    print(f"{title} — {price}")
```

```python
import pandas as pd

all_books = []

for book in books:
    title = book.find('h3').find('a')['title']
    price = book.find('p', class_='price_color').text
    rating = book.find('p', class_='star-rating')['class'][1]
    
    all_books.append({
        'title': title,
        'price': price,
        'rating': rating
    })

df_books = pd.DataFrame(all_books)
print(df_books)
print(f"\nTotal books scraped: {len(df_books)}")
```

```python
df_books['price'] = df_books['price'].str.replace('Â£', '').str.strip().astype(float)

print(df_books)
df_books.to_csv('books_scraped.csv', index=False)
print("saved!")
```

```python
all_books = []

for page in range(1, 6):  # first 5 pages only
    url = f"http://books.toscrape.com/catalogue/page-{page}.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    books = soup.find_all('article', class_='product_pod')
    
    for book in books:
        title = book.find('h3').find('a')['title']
        price = book.find('p', class_='price_color').text
        rating = book.find('p', class_='star-rating')['class'][1]
        
        all_books.append({
            'title': title,
            'price': price,
            'rating': rating
        })
    
    print(f"Page {page} done — {len(books)} books scraped")

print(f"\nTotal: {len(all_books)} books")
```

```python
df_all = pd.DataFrame(all_books)
df_all['price'] = df_all['price'].str.replace('Â£', '').str.strip().astype(float)

print(df_all)
print(f"\nCheapest book: £{df_all['price'].min()}")
print(f"Most expensive: £{df_all['price'].max()}")
print(f"Average price: £{df_all['price'].mean():.2f}")

df_all.to_csv('books_100.csv', index=False)
print("\nSaved to books_100.csv")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.hist(df_all['price'], bins=15, color='red', edgecolor='black')
plt.title('Price Distribution of 100 Books')
plt.xlabel('Price (£)')
plt.ylabel('Number of Books')
plt.show()
```
