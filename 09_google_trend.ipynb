```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt
import time

pytrends = TrendReq(hl='en-US', tz=360)
print("Connected!")
```

```python
keywords = ['Pashmina shawl', 'Kashmiri shawl', 'Kani shawl', 'luxury shawl']

pytrends.build_payload(keywords, timeframe='today 5-y', geo='')
df_trends = pytrends.interest_over_time().drop(columns=['isPartial'])
print("Worldwide done")
print(df_trends.head())
```
